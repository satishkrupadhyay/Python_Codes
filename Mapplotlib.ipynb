```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
#functional method
plt.plot(x,y)
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#object oriented method

fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('Larger plot')
axes2.plot(y,x)

axes2.set_title('Smaller plot')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes1.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[0].set_title('First')
axes[1].plot(y,x)
axes[1].set_title('Second')
```

```python
axes
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1,figsize=(8,2))
axes[0].plot(x,y)

axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig.savefig('my_picture.png', dpi=200)
```

```python
fig = plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x, x**2, label='X Squared')
ax.plot(x, x**3, label='X Cubbed')

ax.legend(loc=0)
```

## Plot Appearance

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color='orange', lw=3, ls='-', marker='o', markersize=10, 
       markerfacecolor='blue', markeredgewidth=3, markeredgecolor='red')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color='purple', lw=2, ls='--')

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
